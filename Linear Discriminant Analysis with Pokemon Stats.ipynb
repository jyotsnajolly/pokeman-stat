```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv('pokemon.csv')
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1.isna().sum()
```

```python
df = df1[df1['type2'].isnull()]
df.shape
```

```python
df1.columns
```

```python
df = df1[df1['type2'].isnull()].loc[:,['sp_attack', 'sp_defense', 'attack','defense','speed','hp','type1']]
```

```python
X = df.iloc[:,:-1].values
```

```python
from sklearn.preprocessing import normalize
X_norm = normalize(X)
```

```python
y = df.iloc[:,-1].values
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
model1 = LinearDiscriminantAnalysis(n_components=3)
model1.fit(X_norm,y)
```

```python
model1.classes_
```

```python
plt.style.use('fivethirtyeight')
import seaborn as sns
```

```python
fig, ax = plt.subplots()
sns.heatmap(pd.DataFrame(model1.coef_,
           columns = df.columns[:-1],
            index = [model1.classes_]),
            ax=ax, cmap='RdBu', annot =True)
plt.suptitle('LDA Feature Coefficients')
pass
```

```python
pd.Series(np.abs(model1.coef_).sum(axis=1), index=model1.classes_).sort_values().plot.bar(figsize =(12,6), title ="LDA Class Coefficient Sums")
```

```python
model1.explained_variance_ratio_
```

```python
X_hat = model1.fit_transform(X,y)
X_hat
```

```python
colors =plt.cm.get_cmap(name = 'tab20').colors
categories = pd.Categorical(pd.Series(y)).categories
ret = pd.DataFrame({'C1':X_hat[:,0], 'C2':X_hat[:,1], 'Type': pd.Categorical(pd.Series(y))})
ret
```

```python
fig, ax = plt.subplots(1, figsize = (12,6))
for col, cat in zip(colors, categories):
    (ret.query('Type ==@cat').plot.scatter(x = 'C1', y= 'C2', color = col, label = cat, ax = ax,
                                          s=100, edgecolor = 'black', linewidth =1,
                                          title = 'Two-Component LDA Decomposition')
                              .legend(bbox_to_anchor = (1.2,1))
    )
```
